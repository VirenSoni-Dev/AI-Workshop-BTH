```python
import tensorflow as tf
import numpy as np
```

```python
#XOR Function Implementation using python lib tensorflows(keras)and numpy
# XOR Data
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0],[1],[1],[0]])

# build model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, activation='relu',input_shape =(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile & train
model.compile(optimizer= 'adam', loss='binary_crossentropy',
              metrics= ['accuracy'] )
model.fit(X,y, epochs=10000,verbose=0)

#Predict
print(model.predict(X.round()))
```

```python
import pandas as pd # pandas is used to read csv file

# to input csv file from url
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

#print first 5 rows of data
print(df.head())
```

```python
# to print info about data
print(df.info())
```

```python
# to print statistical summary of data
print(df.describe())
```

```python
 # to print missing values
print(df.isnull().sum())
```

```python
# to fill missing values with mean
df.fillna(df.select_dtypes(include = 'number').mean(), inplace = True)
```

```python
# to drop duplicate values
df.drop_duplicates(inplace = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x = 'species', data = df)
plt.title('Count of Species')
plt.show()
```
