```python
# Neural Networks
#  A Computational Model Inspired by the human brain
#  that learn patterns from data
# Structure of NN
# 1. Neurons
# 2. Connections
# 3. Layers
# Neurons : Units that takes input and produces output

# Types of Layers:
# 1. Input Layer: Recieves raw data
# 2. Hidden Layer: this is where computation happens
# 3. Output Layer: Produces final output

#Connections: Flow of data between Layers
# through weighted connections

# weighted connection : These weights are learned during the
# training process

#Activation Functions: Mathematical func. in NN that determines whether
# a neuron
# 1. RLU(Rectifier Linear Unit):Max(0, X), Fast and efficient
# 2. Sigmoid(): Input between 0 and 1,  Good for probability
# 3. Tanh():just like sigmoid but varies between -1 to 1,
#  (better gradients)

#  Why Non-linearity is Imp ?
# 1. Without Activation Function, Neural Networks would just be linear
# Transformation, which serverly limits what they can learn

# what does Activation func. do ?
# Activation Function helps the NN to learn complex Patterns or Data
# like clusters, Curves, Irregular data or shapes
#  (not just straight lines)

# What is Deep Learning?
# Deep learning refers to NN with multiple layers and enables them
#  to learn hierarchical representations of data
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
#Stochastic Gradient Descent: Is an optimization Algorithm used in ML
# to train model
```

```python
# creating XOR Dataset
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

#Create Model
model = Sequential()
model.add(Dense(1, activation='sigmoid', input_dim=2))
model.add(Dense(1, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))#changed to 1 unit
```

```python
#compile the model
model.compile(optimizer=SGD(learning_rate=0.03),
              loss='binary_crossentropy',metrics= ['accuracy'])
```

```python
# Train model
model.fit(x,y,epochs=1000,verbose=0)
```

```python

#predict prediction
print("Predictions: ")
print(model.predict(x).round())
```
