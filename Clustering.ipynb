```python
#classification supervised learning : predicts the class,
#in which the input belongs to
#regression supervised learning : predicts a real number based
#on input features

# Unsupervised Learning (Uses Unlabeled data)
# Clustering: Natural Groupings in data by measuring similarity or distance
# K-means: Divides data in "K" Distinct clusters
#  1. to find the number of clusters 'K'
#  2. Send each point to nearest cluster centre
#  3. Updates the centres until it stops changing
# Use cases: Customer segmentation, Document clustering, Market basket analysis
# image compression

# #Dimensionality reduction (PCA)
# reduce the number of input variables while retaining essential information
#PCA( Principal component analysis) :  1. converts features into principal components
 #      2. Captures maximum variance in dimensions
 # uses : noise reduction, visualize high dimension data,
#          preprocessing before clustering and classification
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load Data

iris = load_iris()
x = iris.data

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(x)
labels = kmeans.labels_


# Visualize (first 2 features)
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap = "viridis")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```
