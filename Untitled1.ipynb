```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification
```

```python
# Simulate a dataset with 100 features
X, y = make_classification(n_samples=200, n_features=100 , random_state=0)

# Apply PCA to reduce with 2 dimensions
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

#plot in 2d
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='viridis', s= 40)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Visualization : Reducing 100 features to 2')
plt.grid(True)
plt.show()
```
